```python
# Importing pandas and reading in webs scraped IMDB reviews file for episode 3
import pandas as pd
star_wars_3_df = pd.read_csv("/content/INST737_Star_Wars_IMDB_StarWars3.csv")
```

```python
# Prints the first few columns
star_wars_3_df.head()
```

```python
# Drops unneeded columns 
star_wars_3_df_v2 = star_wars_3_df.drop(columns = ["Review_Date", "Author", "Rating", "Review_Title", "Review_Url", "Unnamed: 0"])
```

```python
# Prints newly edited dataframe
star_wars_3_df_v2
```

```python
# Test block to see if review gets printed
reviewParagraph = ""
for review in star_wars_3_df_v2['Review'][:1]:
  reviewParagraph = review

```

```python
# Prints results from previous test for loop 
reviewParagraph
```

```python
# Imports TextBlob
from textblob import TextBlob
blob = TextBlob(reviewParagraph)
```

```python
# This block was used for testing purposes to see if the first review's sentiment and polarity would get displayed
blob.sentiment
```

```python
# Creates dataframe that will be used to get the sentiment and polarity of all star wars episode 3 reviews
star_wars_episode_3_final_df = pd.DataFrame(columns = ['Review', 'Polarity', 'Subjectivity'])
```

```python
# Checks to see if dataframe was created with the columns created above 
star_wars_episode_3_final_df
```

```python
# This block would give the polarity and subjectivity of all reviews for Star Wars episode 3
star_wars_review_list = []
star_wars_polarity_list = []
star_wars_subjectivity_list = []

for review in star_wars_3_df_v2['Review']:
  blob = TextBlob(str(review))
  star_wars_review_list.append(review)
  star_wars_polarity_list.append(blob.sentiment.polarity)
  star_wars_subjectivity_list.append(blob.sentiment.subjectivity)



star_wars_episode_3_final_df['Review'] = star_wars_review_list
star_wars_episode_3_final_df['Polarity'] = star_wars_polarity_list
star_wars_episode_3_final_df['Subjectivity'] = star_wars_subjectivity_list


star_wars_episode_3_final_df = star_wars_episode_3_final_df[['Review', 'Polarity', 'Subjectivity']]
  
star_wars_episode_3_final_df.head()

  
```

```python
# Converting dataframe to csv 
# Code down below was commented out because file already created 
# star_wars_episode1_final_df.to_csv('Star_Wars_Episode3_Sentiment_Analysis.csv’)
```

```python
# Reviews that have a polarity greater than 0 go to a positive dataframe
positive_df = star_wars_episode_3_final_df[star_wars_episode_3_final_df['Polarity'] > 0]
positive_df
```

```python
# Reviews that have a polarity less than 0 go to a negative dataframe
negative_df = star_wars_episode_3_final_df[star_wars_episode_3_final_df['Polarity'] < 0]
negative_df
```

```python
# Creates a sample dataframe of 5 percent of positive reviews
positive_df_2 = positive_df.sample(frac = .05)
# Creates a sample dataframe of 5 percent of negative reviews
negative_df_2 = negative_df.sample(frac = .05)
```
